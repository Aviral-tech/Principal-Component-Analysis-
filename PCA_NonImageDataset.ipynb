```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import load_breast_cancer

from sklearn.preprocessing import StandardScaler
```

```python
cancer=load_breast_cancer()
```

```python
cancer
```

```python
cancer.keys()
```

```python
df=pd.DataFrame(cancer['data'],columns=cancer['feature_names'])
```

```python
df
```

```python
arr=np.array(df)
arr
```

```python
arr=StandardScaler().fit_transform(arr)
arr_cov = (1 / 568) * arr.T.dot(arr)
arr_cov
```

```python
eigen_values, eigen_vectors=np.linalg.eig(arr_cov)
```

```python
#principal components
pc1=arr.dot(eigen_vectors[:,0])
pc2=arr.dot(eigen_vectors[:,1])
pc3=arr.dot(eigen_vectors[:,2])
```

```python
plt.figure(figsize=(8,6))
plt.scatter(pc1,pc2,c=cancer['target'])
plt.xlabel('First PC')
plt.ylabel('Second Pc')
```

```python
plt.figure(figsize=(8,6))
plt.scatter(pc1,pc3,c=cancer['target'])
plt.xlabel('First PC')
plt.ylabel('Second Pc')
```
